과거와 미래데이터간의 사간적인 의존성을 고려한 예측모델을 평가하기 위한 지표

mse(mean squared error, L2 Loss) 



```python
import numpy as np

y = np.array([10, 20, 30, 40, 50])
y_hat = np.array([30, 40, 50, 60, 70])

mse = np.mean((y - y_hat)**2)
mse
```

```python
rmse = mse**(1/2) # error가 과대해져서 root를 붙이는 것임
rmse
```

mae(mean absolute error, L1 Loss)


```python
mae = np.mean(abs(y-y_hat))
mae
```

mape(mean absolute percentage error)


```python
mape = np.mean( abs((y-y_hat)/y) ) 
mape
```
